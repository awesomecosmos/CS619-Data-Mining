```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('hw4_q2_summary.csv')
print(df.columns)
x = df["training test size"]
```

```python
# plot with various axes scales
plt.figure(figsize=(8, 6), dpi=300)

# linear
# plt.subplot(221,sharey=True)
plt.subplots(1, 2, sharey='row')
plt.plot(x,df["tr_min_acc"],color="dodgerblue",alpha=0.5)
plt.plot(x,df["tr_max_acc"],color="dodgerblue",alpha=0.5)
plt.plot(x,df["tr_mean_acc"],color="r")
plt.grid(True)
plt.ylim(top=101)
plt.xlabel("training test size (%)")
plt.ylabel("training accuracy")
plt.title("Accuracy for Trained Data vs Training Test Size")


# log
# plt.subplot(222)
# plt.subplots(1, 2, sharey='row')
plt.plot(x,df["te_min_acc"],color="dodgerblue",alpha=0.5)
plt.plot(x,df["te_max_acc"],color="dodgerblue",alpha=0.5)
plt.plot(x,df["te_mean_acc"],color="r")
plt.grid(True)
plt.ylim(top=101)
plt.xlabel("testing test size (%)")
plt.ylabel("testing accuracy")
plt.title("Accuracy for Tested Data vs Training Test Size")
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(8, 6), dpi=300)

ax1.plot(x,df["tr_min_acc"],color="dodgerblue",alpha=0.5,label="min")
ax1.plot(x,df["tr_max_acc"],color="limegreen",alpha=0.5,label="max")
ax1.plot(x,df["tr_mean_acc"],color="r",label="mean")
ax1.grid(True)
ax1.set_xlabel("training test size (%)")
ax1.set_ylabel("training accuracy (%)")
ax1.set_title("Trained Data Accuracy")
ax1.legend()

ax2.plot(x,df["te_min_acc"],color="dodgerblue",alpha=0.5)
ax2.plot(x,df["te_max_acc"],color="limegreen",alpha=0.5)
ax2.plot(x,df["te_mean_acc"],color="r")
ax2.grid(True)
ax2.set_xlabel("testing test size (%)")
ax2.set_ylabel("testing accuracy (%)")
ax2.set_title("Tested Data Accuracy")
ax2.legend()
```
